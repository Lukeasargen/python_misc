```python
import matplotlib.pyplot as plt
```

```python
def ceil_divide(n, d):
    return (n+d-1)//d
```

```python
d = 5
x_list = list(range(20))
y1_list = [ceil_divide(x, d) for x in x_list]
y2_list = [x//d for x in x_list]
y3_list = [1 + x//d for x in x_list]
```

```python
plt.plot(x_list, y1_list)
plt.plot(x_list, y2_list)
plt.plot(x_list, y3_list)
```
